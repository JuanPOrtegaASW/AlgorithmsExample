```python
def path_cost(path,child_costs):
    sum = 0
    for node in path:
        state = node.state
        for key, value in child_costs.items():
            if key == tuple(state):
                sum = value + sum
    return sum

class Agent:

    def __init__(self, world, costs, initial_state, goal_state):

        self.moves = np.array([[0,1], [0,-1], [-1,0], [1,0]])
        self.initial_state = initial_state
        self.goal_state = goal_state
        self.current_state = initial_state
        self.world = world
        self.costs = costs

    def possible_moves(self):

        succ_states, succ_actions, succ_costs = [],[],[]

        for move in self.moves:

            # check that the next state is in the range of the world
            #print(self.current_state)
            if 0 <= self.current_state[0] + move[0] < self.world.shape[0] and 0 <= self.current_state[1] + move[1] < self.world.shape[1]:

                # move and assign new_state to the position after moving
                new_state = [self.current_state[0] + move[0], self.current_state[1] + move[1]]

                action = [move[0], move[1]]

                cost = self.costs[self.world[new_state[0], new_state[1]]]

                succ_states.append(new_state)
                succ_actions.append(action)
                succ_costs.append(cost)

        return succ_states, succ_actions, succ_costs


    def set_current_state(self, nex_state):

        self.current_state = nex_state



class Node:

    def __init__(self, state, parent=None, action=None, cost=None):
        """Create a search tree Node, derived from a parent by an action."""
        self.state = state
        self.parent = parent
        self.action = action
        #self.cost = cost

        self.depth = 0
        self.path_cost = 0

        if parent:
            self.path_cost = parent.path_cost + cost
            self.depth = parent.depth + 1


    def successors(self, agent):

        succ_states, succ_actions, succ_costs = agent.possible_moves()
        children = []

        for i in range(len(succ_states)):
            children.append(Node(succ_states[i], self, succ_actions[i], succ_costs[i]))
        return  children

    def path(self):
        """Return a list of nodes forming the path from the root to this node."""
        node, path_back = self, []
        # TODO
        return list(reversed(path_back))


def depth_first_graph_search(agent):
    #TODO
    """
    Search the deepest nodes in the search tree first.
    Search through the successors of a problem to find a goal.
    The argument frontier should be an empty queue.
    Does not get trapped by loops.
    If two paths reach a state, only use the first one.
    """
    cur_node = Node(agent.initial_state)
    cur_path = [cur_node].copy()
    frontier = deque([cur_path].copy())  # Stack
    node_frontier = deque([cur_path[-1]].copy())
    explored  = cur_path.copy()
    explored_paths = [cur_path].copy()
    if(cur_node.state == agent.goal_state): 
        return cur_path
    while frontier:
        cur_path = frontier.pop()
        node_frontier.pop()
        cur_node = cur_path[-1]
        agent.current_state = cur_node.state
        children = [node for node in cur_node.successors(agent) if all(node.state != exp.state for exp in explored)]
        if(len(children)>0):
            for child in children:
                child_path = cur_path + [child]
                if(child.state == agent.goal_state):
                    explored.append(child)
                    explored_paths.append(child_path)
                    return child_path
                if all(child.state != exp.state for exp in explored):
                    #for node in cur_path:
                        #print(node.state)
                    explored_paths.append(child_path)
                    explored.append(child)
                    #for node in child_path:
                        #print(node.state)
                    frontier.append(child_path)
                    node_frontier.append(child)


def breadth_first_graph_search(agent):
    #TODO
    """
    Search through the successors of a problem to find a goal.
    The argument frontier should be an empty queue.
    Does not get trapped by loops.
    If two paths reach a state, only use the first one.
    """
    cur_node = Node(agent.initial_state)
    cur_path = [cur_node].copy()
    frontier = deque([cur_path].copy())  # Stack
    node_frontier = deque([cur_path[-1]].copy())
    explored  = cur_path.copy()
    explored_paths = [cur_path].copy()
    if(cur_node.state == agent.goal_state): 
        return cur_path
    while frontier:
        cur_path = frontier.pop()
        node_frontier.pop()
        cur_node = cur_path[-1]
        agent.current_state = cur_node.state
        children = [node for node in cur_node.successors(agent) if all(node.state != exp.state for exp in explored)]
        if(len(children)>0):
            for child in children:
                child_path = cur_path + [child]
                if(child.state == agent.goal_state):
                    explored.append(child)
                    explored_paths.append(child_path)
                    return child_path
                if all(child.state != exp.state for exp in explored):
                    explored_paths.append(child_path)
                    explored.append(child)
                    frontier.appendleft(child_path)
                    node_frontier.appendleft(child)




def uniform_cost_search(agent):
    #TODO
    """
    Search the deepest nodes in the search tree first.
    Search through the successors of a problem to find a goal.
    The argument frontier should be an empty queue.
    Does not get trapped by loops.
    If two paths reach a state, only use the first one.
    """
    cur_node = Node(agent.initial_state)
    cur_path = [cur_node].copy()
    child_values = {}
    child_values[tuple(cur_node.state)]=0
    frontier = deque([cur_path].copy())  # Stack
    node_frontier = deque([cur_path[-1]].copy())
    explored  = cur_path.copy()
    explored_paths = [cur_path].copy()
    if(cur_node.state == agent.goal_state): 
        return cur_path
    while frontier:
        cur_path = frontier.pop()
        node_frontier.pop()
        cur_node = cur_path[-1]
        agent.current_state = cur_node.state
        children = [node for node in cur_node.successors(agent)]
        if(len(children)>0):
            for child in children:
                child_path = cur_path + [child]
                symbol = agent.world[[child.state[0]][0]][child.state[1]]
                value = path_cost(cur_path,child_values) + agent.costs[symbol]
                child_values[tuple(child.state)] = value 
                if(child.state == agent.goal_state):
                    explored.append(child)
                    explored_paths.append(child_path)
                    return child_path
                if any(child.state == exp.state for exp in explored):
                    del child_values[tuple(child.state)]
                if all(child.state != exp.state for exp in explored):
                    explored_paths.append(child_path)
                    explored.append(child)
                    frontier.appendleft(child_path)
                    node_frontier.appendleft(child)  
            del child_values[tuple(cur_node.state)]
            next_node = list(min(child_values, key=child_values.get)) #encuentro el nodo con distancia manhattan mas chiki
            node_frontier_copy = node_frontier.copy()
            for node in node_frontier_copy:
                if node.state == next_node:
                    node_frontier.remove(node) #elimino el elemento con menor valor para agregarlo en ultima posicion para que sea el siguiente a explorar
                    node_frontier.append(node)
            frontier_copy=frontier.copy()
            for path in frontier_copy:
                last_of_path=path[-1]
                if last_of_path.state==next_node:
                    #print('Next node is', last_of_path.state)
                    frontier.remove(path) #elimino el elemento con menor valor para agregarlo en ultima posicion para que sea el siguiente a explorar
                    frontier.append(path)
            #for node in frontier[-1]:
             #   print(node.state)

def order_frontier(frontier):

    costs = []
    new_list = []

    for i in range(len(frontier)):

        costs.append(frontier[i].path_cost)

    costs = np.array(costs)

    indexes = np.argsort(costs)

    for i in range(len(frontier)):

        new_list.append(frontier[indexes[i]])

    return list(reversed(new_list))



Ada = Agent(map1, COSTS, [0,0], [12,15])
g_node = Ada.goal_state

print(g_node)

p = depth_first_graph_search(Ada)

moves = {'01':'â©','0-1':'âª','-10':'â«', '10':'â¬'}

map_copy = map1.copy()


for i in range(len(p)-1):

    ac = p[i]
    n = p[i+1]

    map_copy[ac.state[0], ac.state[1]] = moves[str(n.action[0])+str(n.action[1])]


for i in range(map_copy.shape[0]):
    print(map_copy[i,:])
```
